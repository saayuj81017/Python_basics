## 1. Quick Warm-Up: `if` and `else`

Before we learn loops, let us quickly review how Python makes decisions.

An `if` statement checks a condition. If the condition is true, one block runs. Otherwise, the `else` block runs.

````python
number = 7

if number % 2 == 0:
    print(number, "is even")
else:
    print(number, "is odd")
````

## 2. Switch-Like Cases in Python

In some languages, students learn a `switch` statement.

In Python, especially beginner-friendly Python, we usually solve this using `if`, `elif`, and `else`.

This gives us a clean way to choose from many cases.

````python
day = 3

if day == 1:
    print("Monday")
elif day == 2:
    print("Tuesday")
elif day == 3:
    print("Wednesday")
else:
    print("Another day")
````

## 3. Why Do We Need Loops?

Sometimes we want to do the same action many times.

Instead of writing `print()` again and again, we can use a loop. A loop repeats a block of code.

````python
print("Hello")
print("Hello")
print("Hello")
````

The code above works, but it is repetitive. A loop gives us a shorter and smarter way.

````python
for i in range(3):
    print("Hello")
````

## 4. `while` Loop

A `while` loop keeps running **while a condition is true**.

Basic idea:

```python
while condition:
    do something
```

A `while` loop usually needs a variable that changes inside the loop.

### Example: Count from 1 to 5

````python
count = 1

while count <= 5:
    print(count)
    count = count + 1
````

What is happening here?

- `count` starts at 1
- the loop runs while `count <= 5`
- after printing, we increase `count` by 1
- when `count` becomes 6, the loop stops

## 5. Another `while` Example

We can also use a `while` loop to count backwards.

````python
num = 5

while num >= 1:
    print(num)
    num = num - 1

print("Blast off!")
````

### Watch Out

If we forget to change the loop variable, the loop may run forever. That is called an **infinite loop**.

````python
# Example only: do not run this
# x = 1
# while x <= 5:
#     print(x)
````

## 6. `for` Loop

A `for` loop is very useful when we know **how many times** we want to repeat something.

Basic idea:

```python
for item in sequence:
    do something
```

For beginners, we often use `range()` with `for` loops.

### Example: Print 1 to 5

````python
for number in range(1, 6):
    print(number)
````

In `range(1, 6)`:

- 1 is the starting value
- 6 is not included
- so the output is 1, 2, 3, 4, 5

### More `range()` Examples

````python
print("range(5):")
for i in range(5):
    print(i)

print("range(2, 6):")
for i in range(2, 6):
    print(i)

print("range(2, 11, 2):")
for i in range(2, 11, 2):
    print(i)
````

## 7. Looping Through a String

````python
for letter in "python":
    print(letter)
````

A `for` loop can also go through items one by one, like letters in a word or values in a list.

## 8. `while` vs `for`

Use a `while` loop when:

- the loop should continue until a condition becomes false
- you are not sure exactly how many times it will run

Use a `for` loop when:

- you want to repeat a fixed number of times
- you want to go through a word, list, or range

## 9. Sample Problem

**Sample:** Print the numbers from 1 to 10 and show whether each number is odd or even.

````python
for number in range(1, 11):
    if number % 2 == 0:
        print(number, "is even")
    else:
        print(number, "is odd")
````

This sample combines a loop with an `if` statement. That is a very common pattern in Python.

## 12. Pattern Printing

Pattern printing is a fun way to practice loops.

We often use:

- one loop for a simple pattern
- nested loops for more complex patterns

### Pattern Example 1: Increasing Stars

````python
for i in range(1, 6):
    print("*" * i)
````

The value of `i` changes like 1, 2, 3, 4, 5. Each time, `"*" * i` makes that many stars.

### Pattern Example 2: Decreasing Stars

````python
for i in range(5, 0, -1):
    print("*" * i)
````

### Pattern Example 3: Number Pattern

````python
for i in range(1, 6):
    print(str(i) * i)
````

## 13. Nested Loops

A nested loop means a loop inside another loop.

This is useful for tables and patterns.

````python
for row in range(1, 4):
    for col in range(1, 5):
        print(row, col, end="   ")
    print()
````

In the example above:

- the outer loop changes the row
- the inner loop changes the column
