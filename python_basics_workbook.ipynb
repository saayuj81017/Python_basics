## 3. Basic Data Types

In Python, different kinds of data have different types.

Here are three important ones:

- `str` for text, like `"hello"`
- `int` for whole numbers, like `7`
- `float` for decimal numbers, like `3.14`

```python
message = "hello"
whole_number = 7
decimal_number = 3.14

print(type(message))
print(type(whole_number))
print(type(decimal_number))
```

The `type()` function tells us what kind of data a value is.

## 4. Math in Python

Python can do arithmetic just like a calculator.

Common operators:

- `+` add
- `-` subtract
- `*` multiply
- `/` divide
- `**` powers

```python
print(2 + 3)
print(10 - 4)
print(6 * 7)
print(8 / 2)
print(2 ** 3)
```

```python
score = 10
score = score + 5
print(score)
```

In the second example, the variable `score` starts as `10`, then we add `5` and save the new result back into `score`.

## 5. Strings

Strings are pieces of text.

You can join strings together and also change how they look.

```python
first_name = "Ada"
last_name = "Lovelace"
full_name = first_name + " " + last_name

print(full_name)
print(full_name.upper())
print(full_name.lower())
```

A few useful string ideas:

- `+` joins text together
- `.upper()` makes text uppercase
- `.lower()` makes text lowercase

```python
word = "python"
print(word[0])
print(word[1])
print(word[-1])
```

This is called **indexing**.

Python starts counting positions at `0`, not `1`.

## 6. Lists

A list stores multiple items in one place.

Lists use square brackets: `[]`.

```python
fruits = ["apple", "banana", "orange"]

print(fruits)
print(fruits[0])
print(fruits[1])
```

```python
fruits.append("mango")
print(fruits)
```

Useful list ideas:

- items are separated by commas
- you can access items by position
- `.append()` adds a new item to the end

## 7. Dictionaries

A dictionary stores information as **key-value pairs**.

This is useful when data has labels.

```python
student = {
    "name": "Maya",
    "grade": 9,
    "favorite_subject": "Science"
}

print(student)
print(student["name"])
print(student["favorite_subject"])
```

In a dictionary:

- a **key** is the label, like `"name"`
- a **value** is the information stored for that label

## 8. Comparisons and Boolean Values

Sometimes we want Python to check if something is true or false.

Boolean values are:

- `True`
- `False`

```python
print(5 > 3)
print(2 == 2)
print(10 < 4)
print(7 != 8)
```

Common comparison operators:

- `==` equal to
- `!=` not equal to
- `>` greater than
- `<` less than

## 9. If Statements

An `if` statement lets your program make decisions.

If a condition is true, Python runs the indented code below it.

```python
temperature = 75

if temperature > 70:
    print("It is a warm day.")
```

```python
age = 15

if age >= 18:
    print("You are an adult.")
else:
    print("You are under 18.")
```

Notice the indentation. In Python, spaces matter.

Indented lines belong to the `if` or `else` block.

## 10. Loops

Loops let us repeat code.

A `for` loop is often used to go through items one at a time.

```python
colors = ["red", "green", "blue"]

for color in colors:
    print(color)
```

```python
for number in range(5):
    print(number)
```

The `range(5)` part gives the numbers `0, 1, 2, 3, 4`.

Again, notice the indentation inside the loop.

## 11. Functions

A function is a reusable block of code.

We define a function with `def`.

```python
def greet():
    print("Hello!")

greet()
```

```python
def greet_person(name):
    print("Hello,", name)

greet_person("Sofia")
greet_person("Jamal")
```

A value given to a function is called an **argument**.

In the example above, `name` is a parameter, and `"Sofia"` is an argument.

## 12. Returning Values

Functions can also send a value back using `return`.

```python
def add_numbers(a, b):
    return a + b

result = add_numbers(4, 5)
print(result)
```

Without `return`, a function can do something.

With `return`, a function can also give a result back.
